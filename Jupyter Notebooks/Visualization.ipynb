```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
from shapely.geometry import Polygon
%matplotlib inline
```

```python
f = "./output.txt"
file = open(f, 'r')
lines = file.readlines()

lines = [item.strip() for item in lines]

Polygon_parts = []
count = 0

fig, ax = plt.subplots(figsize=(16,16))
for element in lines:
    if(len(element)!=0):
        Polygon_parts.append(element)
    elif(len(element)==0):
        count = count + 1
        x = [eval(item.split(" ")[0]) for item in Polygon_parts]
        y = [eval(item.split(" ")[1]) for item in Polygon_parts]
        polygon = Polygon(zip(x, y))
        ax.plot(*polygon.exterior.xy)
        ax.set_aspect('equal')
        Polygon_parts.clear()

plt.savefig(f'./visualisation.png')       
plt.show()
```
