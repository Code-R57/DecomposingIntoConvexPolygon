```python
import numpy as np
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import os

time = []
num_points = []
k = 0
for filename in os.listdir("./timing"):
    filename = filename[:-4]
    k = k + 1
    with open(f"./timing/{filename}.txt") as time_file:
#         print(k)
        time.append(eval(time_file.read().splitlines()[0]))
#         time.append(eval(time_file.read().readline().strip()))
    with open(f"./countries/{filename}") as num_file:
        num_points.append(eval(num_file.read().splitlines()[0]))

print(time)
print(num_points)
```

```python
first_list = num_points
second_list = time

sorted_second_list = [x for _, x in sorted(zip(first_list, second_list))]
print(sorted_second_list)
```

```python

# Generate some sample data
num_points = sorted(num_points)
run_time = sorted_second_list

# Create a figure and axes object
fig, ax = plt.subplots(figsize=(16,16))


# Plot the data as a scatter plot with a line connecting the points
ax.plot(num_points, run_time, '-o', color='red', markersize=8, linewidth=2, label='Running Time')

# Set the x-axis and y-axis limits
ax.set_xlim([0, max(num_points)*1.1])
ax.set_ylim([0, max(run_time)*1.1])

# Add a grid to the plot
ax.grid(True, linestyle='--', alpha=0.5)

# Add labels and title
ax.set_xlabel('Number of Points', fontsize=14, fontweight='bold')
ax.set_ylabel('Running Time (seconds)', fontsize=14, fontweight='bold')
ax.set_title('Running Time vs. Number of Points', fontsize=16, fontweight='bold')

# Add legend
ax.legend(loc='upper left', fontsize=12)

# Display the plot
plt.savefig(f'./timechart.png')
plt.show()
```
